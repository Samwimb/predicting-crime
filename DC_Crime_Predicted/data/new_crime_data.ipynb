```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint
```

```python
query_url = "https://maps2.dcgis.dc.gov/dcgis/rest/services/FEEDS/MPD/MapServer/1/query?where=START_DATE>DATE'2019-08-15 00:00:00'&outFields=CCN,START_DATE,DISTRICT&outSR=4326&f=json"

START_DATE =[]
CCN=[]
DISTRICT =[]

response = requests.get(query_url).json()['features']
for x in response:
    try:
        CCN.append(x['attributes']['CCN'])
        START_DATE.append(x['attributes']['START_DATE'])
        DISTRICT.append(x['attributes']['DISTRICT'])
    except (KeyError, IndexError):
        print(KeyError, IndexError)        
```

```python
# create a data frame from crime data
crime_dict = {
    "START_DATE": START_DATE,
    "CCN": CCN,
    "DISTRICT": DISTRICT
}
crime_df = pd.DataFrame(crime_dict)

crime_df['START_DATE_TRF']=pd.DatetimeIndex(pd.to_datetime(crime_df['START_DATE'], unit='ms'))\
                 .tz_localize('UTC' )\
                 .tz_convert('America/New_York')

crime_df['START_DATE_TRF'] = pd.to_datetime(crime_df['START_DATE_TRF']).dt.date
crime_df
 
```

```python
reduced_crime_df = crime_df[['START_DATE_TRF','CCN','DISTRICT']]
reduced_crime_df = pd.DataFrame(reduced_crime_df) 
reduced_crime_df

reduced_crime_df.groupby(['START_DATE_TRF','DISTRICT']).count()
```
