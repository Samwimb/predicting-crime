```python
import pandas as pd
import numpy as np
```

```python
crime_2016 = pd.read_csv("Crime_Incidents_in_2016.csv")
crime_2017 = pd.read_csv("Crime_Incidents_in_2017.csv")
crime_2018 = pd.read_csv("Crime_Incidents_in_2018.csv")
crime_2019 = pd.read_csv("Crime_Incidents_in_2019.csv")
```

```python
crime_2016.head()
```

```python
crime_total = [crime_2016, crime_2017, crime_2018, crime_2019]
merged_crime = pd.concat(crime_total, sort=False)
```

```python
merged_crime.sort_values(by=['START_DATE'])
```

```python
new = merged_crime["START_DATE"].str.split("-", expand = True)
new.head()
```

```python
# new[0].value_counts()
merged_crime["report_year"] = new[0]
```

```python
merged_crime.head()
```

```python
merged_crime['report_year']= merged_crime['report_year'].astype(int)
```

```python
merged_crime = merged_crime[merged_crime.report_year >= 2016]
```

```python
merged_crime['report_year'].value_counts()
```

```python
# merged_crime['START_DATE']= merged_crime['START_DATE'].astype(str)
```

```python
merged_crime['START_DATE']= merged_crime['START_DATE'].map(lambda x: str(x)[:-14])
```

```python
merged_crime.head()
```

```python
crime_count = merged_crime.groupby('START_DATE')['CCN'].count()
crime_count
```

```python
merged_crime.to_csv("combined_crime.csv", index=False)
```

```python
crime_date_count = merged_crime.groupby('START_DATE')['CCN'].nunique()
crime_date_count_df = pd.DataFrame(crime_date_count)
crime_date_count_df.head()
```

```python
crime_date_count_df.rename(columns={'CCN': 'crime-count'})
```
